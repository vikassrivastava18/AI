```python
import numpy as np
```

```python
a = np.zeros(4)
b = np.zeros((4,))
c = np.random.random_sample(10)
```

```python
print(a)
print(b)
print(c)
print(c.shape)
```

```python
a = np.zeros((4,2))
b = np.zeros((2, 4))
c = np.random.random_sample((4, 3))
```

```python
print(a)
print(b)
print(c)
```

```python
a = np.arange(4)
b = np.arange(4.)
c = np.random.rand(4)
d = np.random.rand(4, 2)
```

```python
print(a)
print(b)
print(c)
print(d)
```

```python
a = np.arange(2, 10)
print(a)
```

```python
a = np.array([1,2,3,4,5])
b = -a
c = np.sum(a)
d = np.mean(a)
e = a ** 2
```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
```

## Vector Operations

```python
X = np.random.rand(4, 3)
w = np.random.rand(3)
print(X)
print(w)
print(np.dot(X, w))
```

## Reshape

```python
a = np.arange(8).reshape(-1, 2)
print(a)
```

## Vector 2-D slicing operations

```python
a = np.arange(20).reshape(-1, 10)

print(a[0, 1:8:2])

print(a[:, 0: 9: 3])

print(a[1, :])
```
