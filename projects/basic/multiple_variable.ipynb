```python
import numpy as np
```

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
Y_train = np.array([460, 232, 178])
```

```python
print(X_train[0])
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
b_init = 785.1811367994083
w_init = np.array([0.39133535, 18.75376741, -53.36032453, -26.42131618])
```

```python
print(w_init.shape)
```

```python
def predict(x, w, b):
    n = x.shape[0]
    p = 0
    for i in range(n):
        p += x[i] * w[i]
    p += b
    return p 
```

```python
x_vec = X_train[2, :]
print(x_vec.shape)

# Make prediction
print(predict(x_vec, w_init, b_init))
```

```python
def predict_vectorized(x, w, b):
    p = np.dot(x, w) + b
    return p
```

```python
print(predict_vectorized(x_vec, w_init, b_init))
print(predict_vectorized(X_train, w_init, b_init))
```

```python
def compute_cost(X, y, w, b):
    """
    Args:
        X (ndarray (m, n)): Data, m example with n features
        y (ndarray (m,)): Target values
        w (ndarray(n,)): model weights
        b (scalar): model bias
    """
    m = X.shape[0]

    cost = 0
    for i in range(m):
        cost += (predict_vectorized(X[i], w, b) - y[i]) ** 2

    return cost / (2 * m)
```

```python
print(compute_cost(X_train, Y_train, w_init, b_init))
```

```python
def compute_cost_vect(X, y, w, b):
    m = X.shape[0]
    
    cost = np.sum((predict_vectorized(X, w, b) - y) ** 2)
    return cost / (2*m)
```

```python
print(compute_cost_vect(X_train, Y_train, w_init, b_init))
```

```python
def compute_gradient(X, y, w, b):
    m, n = np.shape(X)
    dj_dw = np.zeros(n)
    dj_db = 0.

    for i in range(m):
        err = predict(X[i], w, b) - y[i]
        for j in range(n):
            dj_dw[j] += err * X[i][j]
        dj_db += err

    dj_dw /= m
    dj_db /=m
    return dj_dw, dj_db
        
```

```python
dj_dw, dj_db = compute_gradient(X_train, Y_train, w_init, b_init)
print(dj_dw)
print(dj_db)
```

```python
import copy, math

def gradient_descent(X, y, w_init, b_init, alpha, num_iters):
    w = copy.deepcopy(w_init)
    b = b_init

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(X, y, w, b)
        
        w -= alpha * dj_dw
        b -= alpha * dj_db

        if i % 1000 == 0:
            cost = compute_cost_vect(X, y, w, b)
            print(f"Iteration: {i:4d}, Cost: {cost:8.2f}")
    return w, b
```

```python
init_w = np.zeros_like(w_init)
init_b = 0.
iterations = 10000
alpha = 5.0e-7

w_final, b_final = gradient_descent(X_train, Y_train, init_w, init_b, alpha, iterations)
```

```python
m = X_train.shape[0]

for i in range(m):
    prediction = predict(X_train[i], w_final, b_final)
    print(f"Actual label: {Y_train[i]}, Prediction: {prediction}")
```

```python
def zscore_normalize_features(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    
    X_norm = (X - mu) / sigma

    return X_norm
```

```python
X_norm = zscore_normalize_features(X_train)
print(X_norm)
```

```python
init_w = np.zeros_like(w_init)
init_b = 0.
iterations = 10000
alpha = 5.0e-4

w_final, b_final = gradient_descent(X_norm, Y_train, init_w, init_b, alpha, iterations)
```

```python
m = X_train.shape[0]

for i in range(m):
    prediction = predict(X_norm[i], w_final, b_final)
    print(f"Actual label: {Y_train[i]}, Prediction: {prediction}")
```
